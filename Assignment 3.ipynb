```python
#Task 2

vgsales= pd.read_csv('vgsales.csv')
vgsales=vgsales.head(20)

plt.figure(figsize=(40,10))
plt.plot(vgsales['Name'], vgsales['Global_Sales'])
plt.title('Video Game Sales')
plt.xlabel('Name')
plt.ylabel('Global_Sales')
plt.show()
```

```python
#Explanation of Task 2
'''
This graph is called the 'Line Chart'. The first 20 data have been showed here. In this case, Will Sports has the highest global
sales and the amount of global sales is decreasing sharply till Super Mario Bros. Afterwards, the trend of global sales is
decreasing in a very slow rate till the end. Besides, there is a also a noticeable decreasing trend of global sales between 
Duck Hunt and Nintendogs.

'''
```

```python
#Task 3

plt.figure(figsize=(40,10))
plt.bar(vgsales['Name'], vgsales['Global_Sales'])
plt.title('Video Game Sales')
plt.xlabel('Name')
plt.ylabel('Global_Sales')
plt.show()
```

```python
#Explanation of Task 3
'''
This graph is called the 'Bar Chart'. The first 20 data have been showed here. 
In this case, Will Sports has the highest global sales and afterwards,Super Mario Bros has the amount of half of global sales 
of Will Sports. Afterwards, the trend of global sales is decreasing in a very slow rate till the end. But, there is a also a 
noticeable decreasing trend of global sales between Duck Hunt and Nintendogs.

'''
```

```python
#Task 4
vgsales= pd.read_csv('vgsales.csv')
vgsales=vgsales.head(50)
sns.jointplot(x="NA_Sales", y="Global_Sales", data=vgsales, kind="reg")
plt.show()
```
